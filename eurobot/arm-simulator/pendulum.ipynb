```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np

from Pendulum import Pendulum
from plot import Span, SubFigure, compose_plot
```

```python
pendulum = Pendulum(2.0, 0.42)
pendulum.reset(0.3, 0.)
dt = 0.01

angle = [pendulum.q]
pos = [pendulum.position()]
time = [0]
for i in range(1000):
    pendulum.step(dt)
    angle.append(pendulum.q)
    pos.append(pendulum.position())
    time.append(time[-1] + dt)

x = lambda pos: np.asarray(pos).T[:][0]
y = lambda pos: np.asarray(pos).T[:][1]

compose_plot(figsize=(15,5), gridsize=(3,2), plots=[
    SubFigure(Span((0, 0), 1, 3), x(pos), y(pos), "X[m]", "Y[m]"),
    SubFigure(Span((0, 1), 1, 1), time, angle, "Time[s]", "Theta[rad]"),
    SubFigure(Span((1, 1), 1, 1), time, x(pos), "Time[s]", "X[m]"),
    SubFigure(Span((2, 1), 1, 1), time, y(pos), "Time[s]", "Y[m]"),
])
```
